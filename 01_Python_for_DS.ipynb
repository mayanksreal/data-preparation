## Python OOP

```python
class Car:
  wheels = 4
```

```python
mustang = Car()
print(mustang.wheels)
del mustang # delete to free-up space
```

```python
class Car:
  def __init__(self, name, speed, year=2026):
    self.name = name
    self.year = year
    self.speed = speed

mustang = Car("GT 500", 250)
```

```python
class Car:
  def __init__(self, name, speed, year=2026):
    self.name = name
    self.year = year
    self.speed = speed

  def start(self):
    print(self.name + " is Starting")

mustang = Car("GT 500", 250)
mustang.start()
```

Inheritance

```python
class Person:
  def __init__(self, age, name):
    self.age = age
    self.name = name

class Student(Person):
  pass

s1 = Student(20, "Alex")
print(s1.age)
```
